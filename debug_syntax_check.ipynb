# Inspect Terminal Environment
This notebook will detect whether the current shell is PowerShell or Node.js REPL and print environment details.

```python
from pathlib import Path
path = Path('src/App.tsx')
text = path.read_text(encoding='utf-8')
print(f'Target file: {path.resolve()}')
print(f'Total lines: {len(text.splitlines())}')
for i, line in enumerate(text.splitlines()[1170:1186], start=1171):
    print(f'{i}: {line}')
print('---')
for i, line in enumerate(text.splitlines()[1520:1540], start=1521):
    print(f'{i}: {line}')
```

```python
from pathlib import Path
path = Path('src/App.tsx')
text = path.read_text(encoding='utf-8')
pairs = {'(': ')', '{': '}', '[': ']'}
stack = []
errors = []
for i, line in enumerate(text.splitlines(), start=1):
    for ch in line:
        if ch in pairs:
            stack.append((ch, i))
        elif ch in pairs.values():
            if not stack:
                errors.append((i, ch, 'extra closing'))
                continue
            open_ch, open_line = stack[-1]
            if pairs[open_ch] == ch:
                stack.pop()
            else:
                errors.append((i, ch, f'mismatch with {open_ch} from line {open_line}'))
                stack.pop()
print('Stack remaining:', stack[:10])
print('Errors:', errors[:20])
```

```python
import subprocess
import sys
cmd = ['powershell', '-NoProfile', '-Command', 'Get-Location; echo OK']
result = subprocess.run(cmd, capture_output=True, text=True)
print('Return code:', result.returncode)
print('STDOUT:', result.stdout)
print('STDERR:', result.stderr)
```
